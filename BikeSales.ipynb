# Numerical analysis

```python
df['Unit_Cost'].describe()
```

```python
print(f"Average Unit Cost is:{round(df['Unit_Cost'].mean(),2)}")
```

```python
df['Unit_Cost'].plot(kind='box', vert=False, figsize=(14,6))
```

BOX plot is a great way to identify outliers

```python
df['Unit_Cost'].plot(kind='density', figsize=(14,6))
```

Density plots are used to understand the probability density function of a dataset, and whether the data is skewed, bimodal or unimodal.

```python
ax = df['Unit_Cost'].plot(kind='density', figsize=(14,6))
ax.axvline(df['Unit_Cost'].mean(), color='red')
ax.axvline(df['Unit_Cost'].median(), color='green')
```

♟ **axvline:** is a function in data visualization libraries like Matplotlib, typically used in Python, to draw a vertical line on a plot or chart. It is a convenient way to mark specific values, points of interest, or reference lines within a plot


```python
plt.figure(figsize=(14,6))
plt.style.use('ggplot') # to get a styled histogram
plt.hist(df['Unit_Cost'], bins=10)
```

## How to derive number of bins to be used
n = number of observations
Range = maximum value – minimum value
Number of intervals =  √n
Width of intervals =  Range / (Number of intervals)

# Categorical Analysis

```python
df1 = df['Age_Group'].value_counts()
df1
```

```python
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

label = ['Adult', 'Young adults', 'Youth', 'Seniors']

fig, ax = plt.subplots()
ax.pie(df1, labels=label, explode=explode, autopct='%1.1f%%',
       shadow=True, startangle=90)
ax.set_title("Age_group")
```
